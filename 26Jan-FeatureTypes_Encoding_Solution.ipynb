# Lab 1 Solution – Feature Types & Categorical Encoding

Dataset: cybersecurity_attacks.csv

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.feature_extraction import FeatureHasher


df = pd.read_csv('cybersecurity_attacks.csv')
df.head()
```

## Identify Feature Types

```
numerical_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object','bool']).columns.tolist()

print('Numerical features:', numerical_cols)
print('Categorical features:', categorical_cols)
```

## Visualization of Feature Types

```
for col in categorical_cols[:2]:
    df[col].value_counts().head(10).plot(kind='bar', title=f'Top categories: {col}')
    plt.show()

for col in numerical_cols[:2]:
    df[col].hist(bins=20)
    plt.title(f'Histogram: {col}')
    plt.show()

    plt.boxplot(df[col].dropna())
    plt.title(f'Boxplot: {col}')
    plt.show()
```

## Categorical Encoding Methods

```
cat_cols = categorical_cols[:3]
X_cat = df[cat_cols].astype(str)

# One-Hot Encoding
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_ohe = ohe.fit_transform(X_cat)
print("One-hot shape:", X_ohe.shape)

# Ordinal Encoding
ord_enc = OrdinalEncoder()
X_ord = ord_enc.fit_transform(X_cat)
print("Ordinal sample:", X_ord[:5])

# Frequency Encoding
X_freq = X_cat.copy()
for c in cat_cols:
    freq = X_freq[c].value_counts(normalize=True)
    X_freq[c] = X_freq[c].map(freq)
print("Frequency encoded sample:")
print(X_freq.head())

# Hashing Encoding
hasher = FeatureHasher(n_features=16, input_type='dict')
X_hash = hasher.transform(X_cat.to_dict(orient='records')).toarray()
print("Hashing shape:", X_hash.shape)
```

## Pros & Cons

**One-Hot:** + no ordering, interpretable | − high dimensionality

**Ordinal:** + compact | − false ordering

**Frequency:** + compact | − loses identity

**Hashing:** + fixed size, handles unseen | − collisions, low interpretability
